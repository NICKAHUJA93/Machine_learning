```python
import numpy as nm
import matplotlib.pyplot as py
data =10+(40)*nm.random.random((2,1000))
nm.round(data)
print(data)
c1x=data[0,0];c1y=data[1,0]
c2x=data[0,1];c2y=data[1,1]
c3x=data[0,2];c3y=data[1,2]
c4x=data[0,3];c4y=data[1,3]
py.figure(1)
for j in range(40):
    c11x=[];c11y=[];c22x=[];c22y=[];c33x=[];c33y=[];c44x=[];c44y=[]
    for i in range(1000):
        d=[]
        d1=nm.sqrt((data[0,i]-c1x)**2+(data[1,i]-c1y)**2)
        d.append(d1)
        d2=nm.sqrt((data[0,i]-c2x)**2+(data[1,i]-c2y)**2)
        d.append(d2)
        d3=nm.sqrt((data[0,i]-c3x)**2+(data[1,i]-c3y)**2)
        d.append(d3)
        d4=nm.sqrt((data[0,i]-c4x)**2+(data[1,i]-c4y)**2)
        d.append(d4)
        minimum =min(d)
        if d1==minimum:
            c11x.append(data[0,i])
            c11y.append(data[1,i])
            py.plot(data[0,i],data[1,i],'r*')
        elif d2==minimum:
            c22x.append(data[0,i])
            c22y.append(data[1,i])
            py.plot(data[0,i],data[1,i],'b*')                
        elif d3==minimum:
            c33x.append(data[0,i])
            c33y.append(data[1,i])
            py.plot(data[0,i],data[1,i],'c*')
        elif d4==minimum:
            c44x.append(data[0,i])
            c44y.append(data[1,i])
            py.plot(data[0,i],data[1,i],'m*')
    c1x=sum(c11x)/len(c11x)
    c1y=sum(c11y)/len(c11y)
    c2x=sum(c22x)/len(c22x)
    c2y=sum(c22y)/len(c22y)
    c3x=sum(c33x)/len(c33x)
    c3y=sum(c33y)/len(c33y)
    c4x=sum(c44x)/len(c44x)
    c4y=sum(c44y)/len(c44y)
    py.plot(c1x,c1y,'ks')
    py.plot(c2x,c2y,'ks')
    py.plot(c3x,c3y,'ks')
    py.plot(c4x,c4y,'ks')
    py.title('iter'+str(j+1))
    py.pause(0.3)
    py.show()
                         
                                                 
                             
                         
                         
```
