```python
import numpy as nm
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
data= load_iris()
x=data['data']
y=data['target']
#converting the iris data into data frame 
df = pd.DataFrame(data.data,columns=data.feature_names)
print(df.head())

df['species']=pd.Categorical.from_codes(data.target,data.target_names)
print(df)

df['is_train']=nm.random.uniform(0,1,len(df))<=.75
print(df)
#train data will be all true value and test data will be all false value
train_data,test_data = df[df['is_train']==True],df[df['is_train']==False]
print(train_data)
print(len(test_data))
features=df.columns[:4]
print(features)
# Factorize value for train_data convert back to zero
y = pd.factorize(train_data['species'])[0]
print("Factorise value",y)
# njobs represent the number of process
clf=RandomForestClassifier(n_jobs=2,random_state=0)
print(clf)
clf.fit(train_data[features],y)
clf.predict(test_data[features])
# To check the probablity of given data
clf.predict_proba(test_data[features])[0:10]
clf.predict_proba(test_data[features])[10:20]
pred=data.target_names[clf.predict(test_data[features])]
pred[0:]
test_data['species'].head()
#cross tab is similar to confusion matrix
pd.crosstab(test_data['species'],pred,rownames=['Actual Species'],colnames=['PredictedSpecies'])
```
