```python
# Reindexing
import numpy as nm
import pandas as pd
df1=pd.DataFrame(np.random.random((10,4)),index=np.arange(100,110))
print("Data frame 1\n",df1)
df2=pd.DataFrame(np.random.random((6,4)),index=np.arange(100,106))
print("Data frame 2\n",df2)
df_index=df2.reindex_like(df1,method="ffill",limit=2)
#limit will fill only two rows ,ffill is forward fill
print("df index\n",df_index)
print(df_index.dropna())
```

```python
# Merging of two different data frame
#Techniques(Outer ,inner ,left ,right)
#inner join(intersect both data frame and display both data frame)
#outer join will give union
#Right join right id based left table merge
#Left join left id based right table merge
import pandas as pd
d1 = {"pid":[101,102,103,104],"pname":['a','b','c','d'],"price":[200,400,500,600],"pcat":["elect","mob","watch","ha"]};
df_left=pd.DataFrame(d1)
print(df_left)
d2 = {"pid":[101,103,102,104,105],"p_dic":[10,20,30,40,50],"pcolor":["r",'g','b','y','p'],"pcat":["elec","mob","watch","ha","it"]
     }
df_right=pd.DataFrame(d2)
print(df_right)
df_inner_merge=pd.merge(df_left,df_right,on=["pid","pcat"])
#it will take intersection of both pid and pcat and display the results
print("Data frame after inner mere\n",df_inner_merge)
df_outer_merge=pd.merge(df_left,df_right,how="outer",on=["pid","pcat"])
#it will take union of both pid and pcat and display the results
print("Data frame after outer merge\n",df_outer_merge)
df_left_merge=pd.merge(df_left,df_right,how="left",on=["pid"])
#it will take intersection of both pid and pcat and display the results
print("Data frame after left mere\n",df_left_merge)
df_right_merge=pd.merge(df_left,df_right,on=["pid"],how="right")
#it will take intersection of both pid and pcat and display the results
print("Data frame after right mere\n",df_right_merge)
```

```python
# Let's solve an example for merging a data frame for cars
import numpy as nm
import pandas as pd
import matplotlib.pyplot as mat
Manufacturing = {"Car ID":[100,101,102,103,104],"Car Type":['Diesal','Petrol','Electric','Hybrid','CNG'],"Car Color":['Red','Yellow','Orange','Green','Brown']}
data = pd.DataFrame(Manufacturing)
x = data["Car ID"]
y = nm.arange(5)
mat.figure(1)
mat.bar(x,y,color='r' )
print("Manufacturing data\n",data)
Maruti_plant = {"Car ID":[100,101,102,109,114],"Car Model":['WagonR','Swift','Caiz','Swift Dzire','Ertiga']}
data1 = pd.DataFrame(Maruti_plant)
print("Maruti plant data\n",data1)
data_inner_joint = pd.merge(data,data1,on=["Car ID"],how="inner",indicator=True)
print("Data after inner joint..!!\n",data_inner_joint)
x1 = data_inner_joint["Car ID"]
y1 = nm.arange(3)
print("Ploting a figure after inner joint")
mat.figure(2)
mat.bar(x1,y1,color='green')
data_outer_joint = pd.merge(data,data1,on=["Car ID"],how="outer",indicator=True)
print("Data after outer joint..!!\n",data_outer_joint)
x2 = data_outer_joint["Car ID"]
y2 = nm.arange(7)
print("Ploting a figure after outer joint")
mat.figure(3)
mat.bar(x2,y2,color='blue')
data_left_joint = pd.merge(data,data1,on=["Car ID"],how="left",indicator=True)
print("Data after left joint..!!\n",data_left_joint)
x3 = data_left_joint["Car ID"]
y3 = nm.arange(5)
print("Ploting a figure after left joint")
mat.figure(4)
mat.bar(x3,y3,color='yellow')
data_right_joint = pd.merge(data,data1,on=["Car ID"],how="right",indicator=True)
print("Data after right joint..!!\n",data_right_joint)
x4 = data_right_joint["Car ID"]
y4 = nm.arange(5)
print("Ploting a figure after right joint")
mat.figure(5)
mat.bar(x4,y4,color='orange')
mat.show()
```
