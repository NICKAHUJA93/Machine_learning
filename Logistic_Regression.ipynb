```python
from sklearn.datasets import load_digits
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
data =load_digits()
print("Printhing the load digit data sets",data)
print("Shape of target",data.target.shape)
print("image data shape",data.data.shape)
plt.figure(figsize=(20,4))
# To print the image
for index ,(image,label) in enumerate(zip(data.data[0:5],data.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.Accent)
    plt.title("t % d "% label,fontsize=10)
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=.23,random_state=2)
print("Shape of x train data",x_train.shape)
print("shape of x test data",x_test.shape)
# call the Logistic Regresiion function 
lr=LogisticRegression()
# train the model
lr.fit(x_train,y_train)
#Predict the model
print(lr.predict(x_test[0].reshape(1,-1)))
#Predict first 10 value from the data set 
lr.predict(x_test[0:10])
predictor = lr.predict(x_test)
score=lr.score(x_test,y_test)
#To find accuracy we use score model
print("Score",score)
#To create the confusion matrix between y_test and predictor 
cm=metrics.confusion_matrix(y_test,predictor)
print("cm",cm)
plt.figure(figsize=(9,9))
sns.heatmap(cm,fmt=".3f",annot=True,linewidths=.5,square=True)
plt.ylabel("actual data")
plt.xlabel("predicted data")
#Putting a title as score
all_title="accuracy {0} ".format(score)
#changing a title size to 15
plt.title(all_title,size=15)

#More train more the accuracy.......!!!!!
#logistic regression is type of classification that uses(Regression(formula to predict the value) as well as classification(1,0))

# Logisistic regression on breast cancer,Sir sharing one csv
# Logistic Regression was used in the biological sciences in early twentieth century. It was then used in many social science applications. Logistic Regression is used when the dependent variable(target) is categorical.

```

```python
# Applying logistic regression on breast cancer dataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = load_breast_cancer()
print(data)
print("Test data",data.target)
print("data",data.data)
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=.8,random_state=2)
#x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=.23,random_state=2)
print("Train values..\n",x_train,y_train)
print("Test values...\n",x_test,y_test)
print("Shape of x train data...!!",x_train.shape)
print("shape of x test data....!!",x_test.shape)
lr = LogisticRegression()
lr.fit(x_train,y_train)
predict=lr.predict(x_test)
accuracy =lr.score(x_test,y_test)
print("accuracy of model in percentage is !!!!!\n ",accuracy*100)
#Accuracy calculation is correct predicted(diagnal of confusion matrix)/Total values 
cm=metrics.confusion_matrix(y_test,predict)
print("confusion matrix is \n",cm)
plt.figure(figsize=(9,9))
sns.heatmap(cm,fmt=".3f",annot=True,linewidths=.5,square=True)
#annot is true write in the block
plt.ylabel("actual data")
plt.xlabel("predicted data")
#Putting a title as score
all_title="accuracy {0} ".format(accuracy)
#changing a title size to 15
plt.title(all_title,size=15)







```
