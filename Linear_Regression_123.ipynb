```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import sklearn.linear_model
b=load_boston()
print(b)
df_x=pd.DataFrame(b.data,columns=b.feature_names)
df_y=pd.DataFrame(b.target)
df_x.describe()
df_y.describe()
reg=sklearn.linear_model.LinearRegression()
print(reg)
X_train,X_test,Y_train,Y_test =train_test_split(df_x,df_y,test_size=0.10,random_state=4)
reg.fit(X_train,Y_train)
p_value=reg.predict(X_test)
np.mean((p_value - Y_test)**2)
reg.score(X_test,Y_test)
```
