```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyp
import seaborn as sns
from sklearn import preprocessing ,svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
csv_path = 'E:\Data_Science\Pokemon.csv'
pokemon = pd.read_csv(csv_path)
#print(pokemon)
# Replace all Nan with Nipun
print(pokemon.fillna("Nipun"))
#Considering only Total and HP
Pokemon_filter = pokemon[['Total','HP']]
# Renaming the columns
Pokemon_filter.columns=['Total value','Horse power']
print(Pokemon_filter)
data1=sns.lmplot(x="Total value",y="Horse power",data =Pokemon_filter,order = 2, ci = None)

# train the model
x =np.array(Pokemon_filter["Total value"]).reshape(-1,1)
y =np.array(Pokemon_filter["Horse power"]).reshape(-1,1)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
regr = LinearRegression()
regr.fit(X_train, y_train)
# To estimate the predictions 
accuracy =regr.score(X_test, y_test)
print("Accuracy of the model in Percentage",accuracy*100)
y_pred = regr.predict(X_test) 
pyp.scatter(X_test, y_test, color ='b') 
pyp.plot(X_test, y_pred, color ='r') 
pyp.show() 




```
