```python
import pandas as pd
import numpy as np
# series will use for one dimension
data= pd.Series([1,2,3,4])
print(data)
print(type(data))
print("minimum data",min(data))
print("maximum data",max(data))
print("length" ,len(data)) 
#Series is used for 1 dimension in Pandas
data0 =pd.Series({'Name':['Nipun'],'College':[np.NAN]})
print("data 0 in above data is",data0)
#data frame is use for 2 dimensions Pandas
data1 = pd.DataFrame({'Country':['India','Japan','Australia','China','Russia'],'Economy Rank':[1,2,3,4,5]})
print(data1)
print("type of dataframe",type(data1))
#if index is of interger type then check the access to that element
data3 = pd.Series(['Alto','Swift',"Kwid"],index=[100,101,102])
print(data3)
# Accessing of element is done by index
print(data3[101])
print("output for describe function",data3.describe())
```

```python
#Creating a data frame using different data type 
import numpy as np
import pandas as pd
data0 = pd.DataFrame([[1,2,3,4],[9,6,4,5],[0,0,0,0],[10,10,10,10]],index=['a','b','c','d'],columns=['x','y','z','a'])
print(data0)

# creating a data frame from nested dictonary
#primary keys become column and secondary keys become rows
d ={'a':{'a1':10,'a2':20,'a3':30,'a4':40},'b':{'b1':100,'b2':200,'b3':300,'b4':400}}
data1 = pd.DataFrame(d)
print(data1)
```
